#### B1 Task 3: Optional fee or cost structure

If your issue involves **comparing the cost structure of AI versus an
incumbent approach** (e.g. AI credit decisioning vs manual underwriting,
robo-advice vs traditional advice), adapt and run the fee calculator
below to illustrate the economics.


```python
import numpy as np
import matplotlib.pyplot as plt

# TODO: Replace with your own cost parameters.
# Example below: AI-based compliance screening vs manual review.
volume = np.array([100, 500, 1_000, 5_000, 10_000, 50_000])  # items/year
labels = [f'{v:,}' for v in volume]

# Incumbent (manual): fixed overhead + per-item cost
incumbent_fixed = 50_000      # £/yr (staff, licensing)
incumbent_per_item = 50       # £ per item reviewed
incumbent_total = incumbent_fixed + incumbent_per_item * volume

# AI-augmented: higher upfront, near-zero marginal
ai_fixed = 120_000            # £/yr (system, compliance, monitoring)
ai_per_item = 0.50            # £ per item
ai_total = ai_fixed + ai_per_item * volume

# Break-even
breakeven = int(round((ai_fixed - incumbent_fixed) / (incumbent_per_item - ai_per_item)))

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(volume, incumbent_total / 1000, 'o-', label='Incumbent (manual)', linewidth=2, color='steelblue')
axes[0].plot(volume, ai_total / 1000, 's--', label='AI-augmented', linewidth=2, color='coral')
axes[0].axvline(breakeven, color='grey', linestyle=':', linewidth=1.5, label=f'Break-even ≈ {breakeven:,} items')
axes[0].set_xlabel('Annual volume (items)')
axes[0].set_ylabel('Total annual cost (£000s)')
axes[0].set_title('Total cost vs volume')
axes[0].legend(fontsize=9)
axes[0].grid(alpha=0.3)

effective_saving = (incumbent_total - ai_total) / incumbent_total * 100
axes[1].bar(labels, np.clip(effective_saving, 0, None), color='coral', alpha=0.8)
axes[1].set_xlabel('Annual volume')
axes[1].set_ylabel('Cost saving (%)')
axes[1].set_title('Cost saving from AI (% of incumbent cost)')
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].grid(axis='y', alpha=0.3)

plt.suptitle('AI vs Incumbent: Cost Economics (adapt parameters to your application)', fontsize=11)
plt.tight_layout()
plt.show()

print(f"Break-even volume: approximately {breakeven:,} items/year")
print(f"At {volume[-2]:,} items: incumbent costs £{incumbent_total[-2]/1000:.0f}k, AI costs £{ai_total[-2]/1000:.0f}k")
```

#### B2 Task 4: Optional walk-forward validation concept

If your claim involves a **performance claim over time** (e.g. a credit
model, fraud model, trading strategy, or forecasting system), the code
below illustrates the difference between in-sample and walk-forward
(out-of-sample) performance; that gap should make you cautious about
vendor-quoted figures.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
T = 120        # months of data
n_folds = 5
train_size = 60
test_size = 12

# Simulate: true signal is weak (SR 0.3); model overfits in-sample to SR 1.2
true_signal = np.random.normal(0.025, 0.08, T)  # monthly returns, true process

# In-sample: model fit on all data (overfits, looks good)
in_sample_sr = 1.20   # vendor-quoted Sharpe

# Walk-forward: genuine out-of-sample (much weaker)
wf_returns = []
wf_bh = []
for fold in range(n_folds):
    start = fold * test_size
    end = start + test_size
    if end > T:
        break
    # Out-of-sample performance reverts toward true signal
    fold_r = true_signal[start:end] + np.random.normal(0, 0.02, test_size)
    wf_returns.extend(fold_r.tolist())
    wf_bh.extend(true_signal[start:end].tolist())

wf_sr = np.mean(wf_returns) / np.std(wf_returns) * np.sqrt(12)
bh_sr = np.mean(wf_bh) / np.std(wf_bh) * np.sqrt(12)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Panel 1: Sharpe ratio comparison
bars = ['In-sample\n(vendor-quoted)', 'Walk-forward\n(honest)', 'Buy & hold\n(baseline)']
values = [in_sample_sr, wf_sr, bh_sr]
colors = ['#d62728', '#2ca02c', '#7f7f7f']
axes[0].bar(bars, values, color=colors, alpha=0.8, width=0.5)
axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].set_ylabel('Annualised Sharpe ratio')
axes[0].set_title('In-sample vs walk-forward Sharpe')
axes[0].grid(axis='y', alpha=0.3)
for bar, val in zip(axes[0].patches, values):
    axes[0].text(bar.get_x() + bar.get_width() / 2, val + 0.03,
                 f'{val:.2f}', ha='center', fontsize=10, fontweight='bold')

# Panel 2: Fold-by-fold performance
fold_sharpes = []
for fold in range(n_folds):
    start = fold * test_size
    end = start + test_size
    if end > T:
        break
    fr = true_signal[start:end] + np.random.normal(0, 0.02, test_size)
    fold_sharpes.append(np.mean(fr) / np.std(fr) * np.sqrt(12))

axes[1].bar(range(1, len(fold_sharpes) + 1), fold_sharpes,
            color=['#2ca02c' if s > 0 else '#d62728' for s in fold_sharpes], alpha=0.8)
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].axhline(in_sample_sr, color='#d62728', linestyle='--', linewidth=1.5,
                label=f'In-sample SR = {in_sample_sr:.1f}')
axes[1].set_xlabel('Walk-forward fold')
axes[1].set_ylabel('Fold Sharpe ratio')
axes[1].set_title('Out-of-sample performance by period')
axes[1].legend(fontsize=9)
axes[1].grid(axis='y', alpha=0.3)

plt.suptitle(f'The Overfitting Gap: in-sample SR {in_sample_sr:.1f} → walk-forward SR {wf_sr:.2f}', fontsize=11)
plt.tight_layout()
plt.show()

print(f"In-sample Sharpe (vendor-quoted): {in_sample_sr:.2f}")
print(f"Walk-forward Sharpe (out-of-sample): {wf_sr:.2f}")
print(f"Reduction: {(1 - wf_sr/in_sample_sr):.0%}")
print(f"\nKey question for any vendor: 'Is your quoted figure in-sample or walk-forward?'")
```

#### B3 Task 2: Confidence trap assessment

The UKFin+ evidence showed that 4 of 6 LLMs displayed **inverse
calibration**: highest confidence on the least accurate outputs. Run the
simulation below using parameters that reflect your system (or the
illustrative defaults) and assess whether confidence-based routing would
be safe for your application.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n_cases = 200   # TODO: set to your approximate annual decision volume

# TODO: Adjust these to reflect your system.
# calibration_type: 'inverse' (high confidence → low accuracy, as in 4/6 UKFin+ models)
#                   'calibrated' (high confidence → high accuracy, ideal)
calibration_type = 'inverse'  # change to 'calibrated' to see the safe scenario

accuracy = np.random.beta(3, 2, size=n_cases)

if calibration_type == 'inverse':
    confidence = 1 - accuracy + np.random.normal(0, 0.08, size=n_cases)
    title_conf = 'Inverse calibration (as in 4 of 6 UKFin+ models)\nHigh confidence → LOW accuracy'
else:
    confidence = accuracy + np.random.normal(0, 0.08, size=n_cases)
    title_conf = 'Well-calibrated model\nHigh confidence → HIGH accuracy'

confidence = np.clip(confidence, 0.5, 1.0)
ambiguity = np.random.beta(2, 3, size=n_cases)

# Routing thresholds
conf_auto_thresh = 0.85
amb_auto_thresh = 0.30
amb_hybrid_thresh = 0.60

# Routing outcomes
auto_conf = confidence > conf_auto_thresh
auto_amb = ambiguity < amb_auto_thresh
hybrid_amb = (ambiguity >= amb_auto_thresh) & (ambiguity < amb_hybrid_thresh)
human_amb = ambiguity >= amb_hybrid_thresh

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Panel 1: Confidence-based routing
ax = axes[0]
sc = ax.scatter(confidence, accuracy, c=accuracy, cmap='RdYlGn', s=30, alpha=0.6)
ax.axvline(conf_auto_thresh, color='red', linestyle='--', label=f'Auto-approve > {conf_auto_thresh}')
danger = (confidence > conf_auto_thresh) & (accuracy < 0.5)
ax.scatter(confidence[danger], accuracy[danger], s=80, facecolors='none',
           edgecolors='red', linewidths=1.5, zorder=10, label=f'{danger.sum()} dangerous auto-approvals')
ax.set_xlabel('Model confidence')
ax.set_ylabel('Actual accuracy')
ax.set_title(f'Confidence-based routing\n{title_conf}')
ax.legend(fontsize=8)
ax.grid(alpha=0.3)

# Panel 2: Ambiguity-based routing
ax = axes[1]
ax.scatter(ambiguity, accuracy, c=accuracy, cmap='RdYlGn', s=30, alpha=0.6)
ax.axvline(amb_auto_thresh, color='green', linestyle='--', label=f'Auto < {amb_auto_thresh}')
ax.axvline(amb_hybrid_thresh, color='orange', linestyle='--', label=f'Hybrid {amb_auto_thresh}–{amb_hybrid_thresh}')
amb_danger = (ambiguity < amb_auto_thresh) & (accuracy < 0.5)
ax.scatter(ambiguity[amb_danger], accuracy[amb_danger], s=80, facecolors='none',
           edgecolors='red', linewidths=1.5, zorder=10, label=f'{amb_danger.sum()} dangerous auto-approvals')
ax.set_xlabel('Text / input ambiguity score')
ax.set_ylabel('Actual accuracy')
ax.set_title('Ambiguity-based routing\nIndependent of model calibration')
ax.legend(fontsize=8)
ax.grid(alpha=0.3)

# Panel 3: Routing split summary
routing_labels = ['Fully automated', 'Hybrid\n(AI + human)', 'Fully human']
conf_split = [auto_conf.sum(), (~auto_conf).sum(), 0]
amb_split = [auto_amb.sum(), hybrid_amb.sum(), human_amb.sum()]
x = np.arange(3)
w = 0.35
ax = axes[2]
ax.bar(x - w/2, conf_split, w, label='Confidence routing', color='#d62728', alpha=0.7)
ax.bar(x + w/2, amb_split, w, label='Ambiguity routing', color='#2ca02c', alpha=0.7)
ax.set_xticks(x)
ax.set_xticklabels(routing_labels, fontsize=9)
ax.set_ylabel('Cases')
ax.set_title(f'Routing split (n = {n_cases} cases)')
ax.legend(fontsize=9)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("ROUTING SUMMARY:")
print(f"\n  Confidence-based: {auto_conf.sum()} auto-approved, {danger.sum()} with accuracy < 50%")
print(f"  Ambiguity-based:  {auto_amb.sum()} auto-approved, {amb_danger.sum()} with accuracy < 50%")
print(f"\n  If your model is inversely calibrated, confidence-based routing is systematically unsafe.")
print(f"  Change calibration_type to 'calibrated' to see how the picture changes.")
```

## Common pitfalls to avoid

-   **Only describing frameworks, not applying them:** state how the
    framework applies to your specific application, what it reveals, and
    what it cannot tell you.
-   **Generic AI discussion not grounded in your organisation:** every
    paragraph should be anchored to the specific service, system, or
    decision you chose in Part 1.
-   **Treating the scaffold as the main task:** the notebook structures
    your thinking; the report demonstrates your analytical judgement.
-   **Recommendation without limitations:** acknowledge what you do not
    know and what assumptions could break; intellectual honesty is a
    mark of good analysis.
-   **Generic ethical or governance discussion:** connect
    accountability, fairness, and regulatory concerns directly to your
    chosen application and its specific failure modes.
